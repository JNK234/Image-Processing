## Harris Corner Detection Algorithm

```python
gray
```

```python
gray_cp = np.float32(gray)
```

```python
dst = cv2.cornerHarris(src=gray_cp,blockSize=2,ksize=3,k=0.04)
```

```python
plt.imshow(dst)
```

```python
dst = cv2.dilate(dst,None)
```

```python
plt.imshow(dst)
```

```python
flat_chess[dst>0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(flat_chess)
```

```python
gray_real_chess_cp = np.float32(gray_real_chess)

dst = cv2.cornerHarris(src=gray_real_chess_cp,blockSize=2,ksize=3,k=0.04)
```

```python
dst = cv2.dilate(dst,None)
```

```python
real_chess[dst > 0.01*dst.max()] = [255,0,0]
```

```python
plt.imshow(real_chess)
```

## Shi-Tomasi Corner detection

```python
flat_chess = cv2.imread('../Reference/DATA/flat_chessboard.png')
flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2RGB)

real_chess = cv2.imread('../Reference/DATA/real_chessboard.jpg')
real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.cvtColor(flat_chess,cv2.COLOR_BGR2GRAY)
gray_real_chess = cv2.cvtColor(real_chess,cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess,64,0.01,10 )
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess,100,0.01,10)

corners = np.int0(corners)
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess,(x,y),3,(255,0,0),-1)
```

```python
plt.imshow(real_chess)
```
