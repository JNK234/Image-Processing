```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
plate = cv2.imread('../Reference/DATA/car_plate.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,10))
    ax = plt.subplot(111)
    plt.imshow(img)
```

```python
display(plate)
```

# Detect the number plate and draw rectangle around it

```python
car_plate = cv2.CascadeClassifier('../Reference/DATA/haarcascades/haarcascade_licence_plate_rus_16stages.xml')
```

```python
def detect_plate(img):
    plate = img.copy()
    
    plate_rects = car_plate.detectMultiScale(plate,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in plate_rects:
        cv2.rectangle(plate,(x,y),(x+w,y+h),(255,0,0),3)
    
    return plate
```

```python
result = detect_plate(plate)
display(result)
```

# Blur the Detected number plate

```python
display(result)
```

```python
def blur_plate(img):
    plate = img.copy()
    
    plate_rect = car_plate.detectMultiScale(plate)
    
    for (x,y,w,h) in plate_rect:
        req = plate[y:y+h,x:x+w]
        req = cv2.medianBlur(req,17)
        plate[y:y+h,x:x+w] = req
    
    return plate
```

```python
res = blur_plate(plate)
```

```python
display(res)
```
