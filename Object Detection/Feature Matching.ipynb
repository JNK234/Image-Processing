```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(img,cmap='gray'):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
reeses = cv2.imread('../Reference/DATA/reeses_puffs.png',0)
```

```python
show_img(reeses)
```

```python
cerels = cv2.imread('../Reference/DATA/many_cereals.jpg',0)
```

```python
show_img(cerels)
```

## Brute Force Detector with ORB

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 = orb.detectAndCompute(reeses,None)
kp2,des2 = orb.detectAndCompute(cerels,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches
```

```python
matches[0].distance
```

```python
len(matches)
```

```python
matches = sorted(matches,key=lambda x:x.distance)
```

```python
reeses_matches = cv2.drawMatches(reeses,kp1,cerels,kp2,matches[:25],None,flags=2)
```

```python
show_img(reeses_matches)
```
