```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../Reference/DATA/sammy_face.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(img,threshold1=200,threshold2=255)
plt.imshow(edges)
```

We can find the appropriate edges of the images using canny edge detection by manipulating threshold1 and threshold2 values

```python
med_val = np.median(img)
```

```python
med_val
```

```python
lower = int(max(0,0.7*med_val))
upper = int(min(255,1.3*med_val))
```

```python
lower , upper
```

```python
edges = cv2.Canny(img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

If the above method doesnt work we should start blurring the image

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
plt.imshow(blurred_img)
```

```python
edges = cv2.Canny(blurred_img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
