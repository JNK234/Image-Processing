```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('../maze00.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
show_pic(img)
```

```python
img.shape
```

```python
maze = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
ret, maze = cv2.threshold(maze,127,255,cv2.THRESH_BINARY_INV)
```

```python
maze.shape
```

```python
show_pic(maze)
```

```python
maze.shape
```

```python
kernel = np.ones((3,3),dtype=np.uint8)
```

```python
open = cv2.morphologyEx(maze,cv2.MORPH_OPEN,kernel)
```

```python
show_pic(open)
```

```python
maze
```

```python
maze = cv2.dilate(maze,kernel,iterations=1)
maze = cv2.erode(maze,kernel,iterations=1)
```

```python
show_pic(maze)
```
